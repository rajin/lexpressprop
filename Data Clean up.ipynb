```python
def area_conversion(in_val):
    try:
        split_val = in_val.split(' ',1)
        if len(split_val) > 1:
            if split_val[1] == 'm²':
                out_val = float(split_val[0])
            elif split_val[1] == 'sq ft':
                out_val = float(split_val[0])*0.092903
            elif split_val[1] == 'Perche(s)':
                out_val = float(split_val[0])*42.21
            elif split_val[1] == 'Toise(s)':
                out_val = float(split_val[0])*3.80
            elif split_val[1] == 'Hectare(s)':
                out_val = float(split_val[0])*10000
            else:
                print('unknown units', in_val)
        elif in_val == 'N.S':
            out_val = np.nan
        return np.round(out_val,1)
    except:
        #print(in_val)
        return np.nan
```

```python
allhouses1['Interior Surface New']=allhouses1['Interior Surface'].apply(area_conversion)
```

```python
allhouses1['Land Surface New']=allhouses1['Land Surface'].apply(area_conversion)
```

```python
def remove_feature_headings(in_val):
    try:
        new_list = ','.join(in_val.split('###')).split(',')
        new_list = [x.strip() for x in new_list]
        if 'General' in new_list:
            new_list.remove('General')
        if 'Interior' in new_list:
            new_list.remove('Interior')
        if 'Exterior' in new_list:
            new_list.remove('Exterior')
        return ','.join(new_list)
    except:
        return in_val
```

```python
allhouses1['features New']=allhouses1['features'].apply(remove_feature_headings)
```
